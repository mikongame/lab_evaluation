# Unsupervised Learning Model Evaluation Lab

Complete the exercises below to solidify your knowledge and understanding of unsupervised learning model evaluation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import AgglomerativeClustering

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning) 
```

Load the wine dataset using sklearn and get the datasets X and y containing the target and the rest of the variables

```python
from sklearn import datasets

wine = datasets.load_wine()

X = wine["data"]
y = wine["target"]
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
kmeans_8 = KMeans(n_clusters=8, random_state=42)
clusters_8 = kmeans_8.fit(X)
score = silhouette_score(X, clusters_8.labels_, metric="euclidean")
score
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans_5 = KMeans(n_clusters=5, random_state=42)
clusters_5 = kmeans_5.fit(X)
score = silhouette_score(X, clusters_5.labels_, metric="euclidean")
score
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans_3 = KMeans(n_clusters=3, random_state=42)
clusters_3 = kmeans_3.fit(X)
score = silhouette_score(X, clusters_3.labels_, metric="euclidean")
score
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
plt.figure(figsize=(18,10))

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,15))
visualizer.fit(X)
visualizer.poof()
visualizer.show()

sns.set_context("talk")
plt.show()
```

```python
plt.figure(figsize=(18,10))

model = KMeans()
visualizer = KElbowVisualizer(model, metric="silhouette", k=(2,15))
visualizer.fit(X)
visualizer.poof()
visualizer.show()

sns.set_context("talk")
plt.show()
```

```python
plt.figure(figsize=(18,10))

model = KMeans()
visualizer = KElbowVisualizer(model, metric="calinski_harabasz", k=(2,15))
visualizer.fit(X)
visualizer.poof()
visualizer.show()

sns.set_context("talk")
plt.show()
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X)

plt.figure(figsize=(18,10))

pca = PCA(n_components = 2)
components = pca.fit_transform(X_train_scaled)
components_df = pd.DataFrame(data=components, columns = ["PC1", "PC2"])
components_df["outcome"] = y

sns.relplot(x="PC1", y="PC2", hue=X["labels"], data=components_df, palette="coolwarm")

sns.set_context("talk")
plt.show()
```
